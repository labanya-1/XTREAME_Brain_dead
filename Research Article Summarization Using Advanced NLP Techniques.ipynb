<h1>2. Import Dependencies</h1>

```python
import os
import torch
import numpy as np
from datasets import load_dataset
from evaluate import load as load_metric
from transformers import (
    AutoTokenizer, AutoModelForSequenceClassification, AutoModelForSeq2SeqLM,
    Trainer, TrainingArguments, DataCollatorWithPadding, DataCollatorForSeq2Seq
)
from rouge_score import rouge_scorer
from sacrebleu import corpus_bleu
from bert_score import score as bert_score

# Optimize PyTorch settings
torch.backends.cudnn.benchmark = True
device = "cuda" if torch.cuda.is_available() else "cpu"

# Load evaluation metrics
rouge = load_metric("rouge")
```

<h1>3. Load Dataset </h1>

```python
# Define dataset path
dataset_path = "/content/drive/MyDrive/Brain_Dead_CompScholar_Dataset.csv"

# Load dataset
dataset = load_dataset("csv", data_files=dataset_path, split="train")
print("Dataset Columns:", dataset.column_names)

# Split dataset into training and validation sets
dataset = dataset.train_test_split(test_size=0.1)
train_dataset, val_dataset = dataset["train"], dataset["test"]
```

<h1>4. Tokenization </h1>

```python
# Load tokenizers
bert_tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")
bart_tokenizer = AutoTokenizer.from_pretrained("facebook/bart-large-cnn")

# Tokenization function for BERT (Extractive)
def extractive_preprocess_function(examples):
    inputs = bert_tokenizer(examples["Document"], padding="max_length", truncation=True, max_length=512)
    labels = [1 if "important" in doc.lower() else 0 for doc in examples["Document"]]  # Sample label strategy
    inputs["labels"] = labels
    return inputs

# Tokenization function for BART (Abstractive)
def abstractive_preprocess_function(examples):
    inputs = bart_tokenizer(examples["Document"], padding="max_length", truncation=True, max_length=1024)
    labels = bart_tokenizer(examples["Summary"], padding="max_length", truncation=True, max_length=256)
    inputs["labels"] = labels["input_ids"]
    return inputs

# Apply tokenization
train_dataset = train_dataset.map(extractive_preprocess_function, batched=True)
val_dataset = val_dataset.map(extractive_preprocess_function, batched=True)
train_dataset_bart = train_dataset.map(abstractive_preprocess_function, batched=True)
val_dataset_bart = val_dataset.map(abstractive_preprocess_function, batched=True)
```

<h1>5. Model Initialization</h1>

```python
# Load Models
extractive_model = AutoModelForSequenceClassification.from_pretrained("bert-base-uncased", num_labels=2).to(device)
abstractive_model = AutoModelForSeq2SeqLM.from_pretrained("facebook/bart-large-cnn").to(device)

# Data Collators
data_collator = DataCollatorWithPadding(tokenizer=bert_tokenizer)
seq2seq_collator = DataCollatorForSeq2Seq(tokenizer=bart_tokenizer, model=abstractive_model)
```

<h1>6. Training</h1>

```python
# Training Arguments for Extractive Model
training_args = TrainingArguments(
    output_dir="./bert_results",
    evaluation_strategy="epoch",
    per_device_train_batch_size=8,
    per_device_eval_batch_size=8,
    num_train_epochs=3,
    save_total_limit=2,
    save_strategy="epoch",
    fp16=True,
    gradient_accumulation_steps=4,
    logging_dir="./logs",
)

# Train Extractive Model
bert_trainer = Trainer(
    model=extractive_model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=val_dataset,
    tokenizer=bert_tokenizer,
    data_collator=data_collator,
)
bert_trainer.train()
extractive_model.save_pretrained("./fine_tuned_bert")

# Training Arguments for Abstractive Model
bart_training_args = TrainingArguments(
    output_dir="./bart_results",
    evaluation_strategy="epoch",
    per_device_train_batch_size=4,
    per_device_eval_batch_size=4,
    num_train_epochs=3,
    save_total_limit=2,
    save_strategy="epoch",
    fp16=True,
    gradient_accumulation_steps=4,
    logging_dir="./logs",
)

# Train Abstractive Model
bart_trainer = Trainer(
    model=abstractive_model,
    args=bart_training_args,
    train_dataset=train_dataset_bart,
    eval_dataset=val_dataset_bart,
    tokenizer=bart_tokenizer,
    data_collator=seq2seq_collator,
)
bart_trainer.train()
abstractive_model.save_pretrained("./fine_tuned_bart")
```

<h1>7. Summarization Functions</h1>

```python
# Load fine-tuned models
extractive_model = AutoModelForSequenceClassification.from_pretrained("./fine_tuned_bert").to(device)
abstractive_model = AutoModelForSeq2SeqLM.from_pretrained("./fine_tuned_bart").to(device)

# Extractive Summarization with Sliding Window
def extractive_summary(text, window_size=512, stride=256):
    sentences = text.split(". ")  # Basic sentence segmentation
    extracted_sentences = []

    for i in range(0, len(sentences), stride):
        chunk = ". ".join(sentences[i:i + window_size])  # Create overlapping chunks
        inputs = bert_tokenizer(chunk, return_tensors="pt", truncation=True, max_length=512).to(device)
        outputs = extractive_model(**inputs)
        logits = outputs.logits
        if torch.argmax(logits) == 1:
            extracted_sentences.append(chunk)  # Keep important sentences

    return " ".join(extracted_sentences) if extracted_sentences else text

# Abstractive Summarization with Sliding Window
def abstractive_summary(text, window_size=1024, stride=512):
    sentences = text.split(". ")
    summaries = []

    for i in range(0, len(sentences), stride):
        chunk = ". ".join(sentences[i:i + window_size])
        inputs = bart_tokenizer(chunk, return_tensors="pt", truncation=True, max_length=1024).to(device)
        summary_ids = abstractive_model.generate(
            inputs["input_ids"], max_length=200, min_length=50, num_beams=5, early_stopping=True
        )
        summaries.append(bart_tokenizer.decode(summary_ids[0], skip_special_tokens=True))

    return " ".join(summaries)

# Hybrid Summarization (Extractive + Abstractive)
def hybrid_summary(text):
    extracted_text = extractive_summary(text)  # Extract key sentences
    return abstractive_summary(extracted_text) if extracted_text else abstractive_summary(text)
```

<h1>8. Evaluation Metrics</h1>

```python
# Evaluation Functions
def evaluate_summary(reference, generated):
    scorer = rouge_scorer.RougeScorer(['rouge1', 'rouge2', 'rougeL'], use_stemmer=True)
    return scorer.score(reference, generated)

def evaluate_bleu(reference, generated):
    return corpus_bleu([generated], [[reference]]).score

def evaluate_bert_score(reference, generated):
    P, R, F1 = bert_score([generated], [reference], lang="en")
    return {"Precision": P.mean().item(), "Recall": R.mean().item(), "F1": F1.mean().item()}

# Sample evaluation
reference = "The proposed model improves accuracy. Experiments confirm its effectiveness."
generated = hybrid_summary(reference)

# Evaluate
rouge_scores = evaluate_summary(reference, generated)
bleu_score = evaluate_bleu(reference, generated)
bert_scores = evaluate_bert_score(reference, generated)

# Print evaluation results
print("Generated Summary:", generated)
print("ROUGE Scores:", rouge_scores)
print("BLEU Score:", bleu_score)
print("BERTScore:", bert_scores)
```
